```python
import numpy as np

# set up matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

from astropy.io import fits
from scipy.ndimage import gaussian_filter
```

```python
from astropy.utils.data import download_file
image_file = download_file('http://data.astropy.org/tutorials/FITS-images/HorseHead.fits', cache=True)
```

**Opening FITS files and loading the image data**

```python
hdu_list = fits.open(image_file)
hdu_list.info()
```

```python
image_data = hdu_list[0].data
```

```python
print(type(image_data))
print(image_data.shape)
```

```python
hdu_list.close()
```

**Shortcut**

```python
image_data = fits.getdata(image_file)
```

```python
print(type(image_data)) # show the python type for image_data
print(image_data.shape) #
```

**Viewing the image data and getting basic statistics**

```python
image_data_smooth = gaussian_filter(image_data, sigma=1)

plt.imshow(image_data, cmap='ocean', vmin=-100, vmax=22918, interpolation='bicubic')
plt.colorbar()
plt.savefig('high_res_image.png', dpi=300)
plt.show()
```

```python
print('Min:', np.min(image_data))
print('Max:', np.max(image_data))
print('Mean', np.mean(image_data))
print('Stdev:', np.std(image_data))
```
